### Looking at the distributions of the data
Let's take a look at the distribution of the 4 stats that describe the physical characteristics of the players.

**Plot the four distributions about `height`, `weight`, `age` and `BMI`.**

```python
#your code here
# fig, axes = plt.subplots(4, 1, figsize=(5, 10))

# sns.distplot(wnba.Height, kde=False, ax=axes[0]) 
# axes[0].set_title('Height')

# sns.distplot(wnba.Weight, kde=False, ax=axes[1])
# axes[1].set_title('Weight')

# sns.distplot(wnba.Age, kde=False, ax=axes[2])
# axes[2].set_title('Age')

# sns.distplot(wnba.BMI, kde=False, ax=axes[3])
# axes[3].set_title('BMI')

# plt.tight_layout()
```

```python
fig, axes = plt.subplots(4, 1, figsize=(5, 12))

wnba.Height.hist(ax=axes[0]) 
axes[0].set_title('Height')

wnba.Weight.hist(ax=axes[1])
axes[1].set_title('Weight')

wnba.Age.hist(ax=axes[2])
axes[2].set_title('Age')

wnba.BMI.hist(ax=axes[3])
axes[3].set_title('BMI')

plt.tight_layout()
```

**What conclusions do you think we can take from this plots?**

```python
#your conclusions here

# all the distributions look like a standard distribution
# the height distribution is skewed to the left, so mostly tall people but also quite a 
# part of the team is a lot smaller (probably the need for fast and small people)

# the age distribution is skewed to the right, which also makes sense as mostly top sport is 
# played by people untill the age of mid 30s

# the average BMI states normal weight untill 25, however, in the USA this is a bit higher
# so therefore, the relatively high distribution from 25 to 26 can be explained
```

In addition to what the describe() method already told us we can see that the physical characteristics of the players (apart from the age) more or less follow a standard distribution, which is expected when looking at the distribution of values that arise from a lot of factors that are independent from each other as is the case for many physical quantities. 

The height distribution looks like a bimodal. This may be due to the fact that players in basketball fall under two main categories (please note that this is a very gross generalization): shorter and more agile, and taller and less agile. Therefore there are less "average" height players since they will neither be as agile as the shorter players nor have the same impact in the paint (that is, under the basket) as a taller player.

The age distribution is a bit skewed to the right which is expected since most professional players stop playing after their prime physical years come to an end.

We could do the same for the main game stats. They are points, assists, blocks, rebounds and steals. 

**Now plot the distribution of the columns `REB`, `AST`, `STL`, `PTS` and `BLK` the same way you did in the last cell.**

```python
#your code here
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(5, 1, figsize=(5, 12))

wnba.REB.hist(ax=axes[0]) 
axes[0].set_title('Total Rebounds')

wnba.AST.hist(ax=axes[1])
axes[1].set_title('Assists')

wnba.STL.hist(ax=axes[2])
axes[2].set_title('Steals')

wnba.PTS.hist(ax=axes[3])
axes[3].set_title('Total Points')

wnba.BLK.hist(ax=axes[4])
axes[4].set_title('Blocks')

plt.tight_layout()
```

**What conclusions do you think we can take from this plots?**

```python
#your conclusions here
# all distributions are skewed to the right, showing all the outliers of these variables on the right side, probably
# showing values of their top players
```

As expected all of the above distribution are heavily skewed to the right, since most players will have very low to average stats while there will be a handful of star players whose stats peak above everyone else. It is also important to think about the fact that we are simply taking the stats as they are without considering the minutes played by each player. Even though skill plays a very important factor in determining this kind of stats we also have to consider that players that play more minutes will, on average, score more points (or blocks, assists, etc.).

**For the sake of it let's look at the same distributions by dividing those stats by the minutes played for each player in the dataset.** 

```python
#your code here
wnba['REB_m'] = wnba['REB'] / wnba['MIN'] 
wnba['AST_m'] = wnba['AST'] / wnba['MIN']
wnba['STL_m'] = wnba['STL'] / wnba['MIN']
wnba['PTS_m'] = wnba['PTS'] / wnba['MIN']
wnba['BLK_m'] = wnba['BLK'] / wnba['MIN']

fig, axes = plt.subplots(5, 1, figsize=(5, 12))

wnba.REB_m.hist(ax=axes[0]) 
axes[0].set_title('Total Rebounds')

wnba.AST_m.hist(ax=axes[1])
axes[1].set_title('Assists')

wnba.STL_m.hist(ax=axes[2])
axes[2].set_title('Steals')

wnba.PTS_m.hist(ax=axes[3])
axes[3].set_title('Total Points')

wnba.BLK_m.hist(ax=axes[4])
axes[4].set_title('Blocks')

plt.tight_layout()
```

**What conclusions do you think we can take from this plots?**

```python
#your conclusions here
# the plots are still skewed to the left, but a lot less, only for a bit, it looks way more like a normal distribution
# this shows that if you divide it by minutes, the outliers become less strong, so the star players with the highest
# score of these columns, is influenced by that they also play the most minutes
```
